```python
import json
import requests



def LoadAuthor(topic, date, pages):
    answer = []
    for j in range(1, pages + 1):    
        url = f"https://api.openalex.org/works?filter=concepts.id:{topic},from_publication_date:{date}&per-page=200&page={j}&"
        response = requests.get(url)
        json_responce = response.json()
        answer_list = json_responce['results']
        for i in range(len(answer_list)):
            authors = []
            for x in range(len(answer_list[i]['authorships'])):
                authors.append(answer_list[i]['authorships'][x]['author']['id'])
            answer.append([answer_list[i]['id'], answer_list[i]['display_name'], answer_list[i]['abstract_inverted_index'], authors])
    with open('data.json', 'w+') as f:
        json.dump(answer, f)
        
```

```python
LoadAuthor('https://openalex.org/C204321447', '2022-01-01', 5)
```
